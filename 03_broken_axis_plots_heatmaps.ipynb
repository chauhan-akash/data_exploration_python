## 1. Plotting Broken Axis charts
### Get the data

```python
set7 = pd.read_csv("data/set7_data.csv")
set7
```

### Default chart
This following plot has outlier values which makes the whole chart unreadable. We will need to break the y axis to show the data clearly

```python
# plot parameter
fig = plt.figure(figsize=(18,6))
plt.rcParams.update({'font.size': 14})

xlab = np.array(set7['Genes'])
day_1 = np.array(set7['0'])
day_2 = np.array(set7['10'])
day_3 = np.array(set7['20'])
day_4 = np.array(set7['30'])
day_5 = np.array(set7['40'])
day_6 = np.array(set7['50'])

day_1_sd = np.array(set7['0_sd'])
day_2_sd = np.array(set7['10_sd'])
day_3_sd = np.array(set7['20_sd'])
day_4_sd = np.array(set7['30_sd'])
day_5_sd = np.array(set7['40_sd'])
day_6_sd = np.array(set7['50_sd'])

# set width of bars
barWidth = 0.1

# Set position of bar on X axis
r1 = np.arange(len(xlab))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
r6 = [x + barWidth for x in r5]

# Make the plot
plt.bar(r1, day_1, yerr = day_1_sd , width=barWidth, edgecolor='white', label='Cycle 0')
plt.bar(r2, day_2, yerr = day_2_sd , width=barWidth, edgecolor='white', label='Cycle 10')
plt.bar(r3, day_3, yerr = day_3_sd , width=barWidth, edgecolor='white', label='Cycle 20')
plt.bar(r4, day_4, yerr = day_4_sd , width=barWidth, edgecolor='white', label='Cycle 30')
plt.bar(r5, day_5, yerr = day_5_sd , width=barWidth, edgecolor='white', label='Cycle 40')
plt.bar(r6, day_6, yerr = day_6_sd , width=barWidth, edgecolor='white', label='Cycle 50')

# Add xticks on the middle of the group bars
plt.xlabel('Genes')
plt.ylabel('Relative Expression')
# plt.title('Relative Expression of Genes by days')
plt.xticks([r + (2*barWidth) for r in range(len(xlab))], xlab)

# Create legend & Show graphic
plt.legend(loc='upper right', prop={'size':10})
plt.show()
```

### Broken Axis chart
In this chart we will break the y axis to show all the values clearly

```python
# plot parameters
fig = plt.figure(figsize=(18,8))
plt.rcParams.update({'font.size': 14})
fig.subplots_adjust(hspace=0.1)  # adjust space between axes
# set height ratios for subplots
gs = gridspec.GridSpec(3, 1, height_ratios=[1, 1, 2]) 

xlab = np.array(set7['Genes'])
day_1 = np.array(set7['0'])
day_2 = np.array(set7['10'])
day_3 = np.array(set7['20'])
day_4 = np.array(set7['30'])
day_5 = np.array(set7['40'])
day_6 = np.array(set7['50'])

day_1_sd = np.array(set7['0_sd'])
day_2_sd = np.array(set7['10_sd'])
day_3_sd = np.array(set7['20_sd'])
day_4_sd = np.array(set7['30_sd'])
day_5_sd = np.array(set7['40_sd'])
day_6_sd = np.array(set7['50_sd'])

# set width of bars
barWidth = 0.1

# Set position of bar on X axis
r1 = np.arange(len(xlab))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
r6 = [x + barWidth for x in r5]


# Make the plot
ax1 = plt.subplot(gs[0])
ax1.bar(r1, day_1, yerr = day_1_sd , width=barWidth, edgecolor='white', label='Cycle 0')
ax1.bar(r2, day_2, yerr = day_2_sd , width=barWidth, edgecolor='white', label='Cycle 10')
ax1.bar(r3, day_3, yerr = day_3_sd , width=barWidth, edgecolor='white', label='Cycle 20')
ax1.bar(r4, day_4, yerr = day_4_sd , width=barWidth, edgecolor='white', label='Cycle 30')
ax1.bar(r5, day_5, yerr = day_5_sd , width=barWidth, edgecolor='white', label='Cycle 40')
ax1.bar(r6, day_6, yerr = day_6_sd , width=barWidth, edgecolor='white', label='Cycle 50')

ax2 = plt.subplot(gs[1], sharex = ax1)
ax2.bar(r1, day_1, yerr = day_1_sd , width=barWidth, edgecolor='white', label='Cycle 0')
ax2.bar(r2, day_2, yerr = day_2_sd , width=barWidth, edgecolor='white', label='Cycle 10')
ax2.bar(r3, day_3, yerr = day_3_sd , width=barWidth, edgecolor='white', label='Cycle 20')
ax2.bar(r4, day_4, yerr = day_4_sd , width=barWidth, edgecolor='white', label='Cycle 30')
ax2.bar(r5, day_5, yerr = day_5_sd , width=barWidth, edgecolor='white', label='Cycle 40')
ax2.bar(r6, day_6, yerr = day_6_sd , width=barWidth, edgecolor='white', label='Cycle 50')

ax3 = plt.subplot(gs[2], sharex = ax1)
ax3.bar(r1, day_1, yerr = day_1_sd , width=barWidth, edgecolor='white', label='Cycle 0')
ax3.bar(r2, day_2, yerr = day_2_sd , width=barWidth, edgecolor='white', label='Cycle 10')
ax3.bar(r3, day_3, yerr = day_3_sd , width=barWidth, edgecolor='white', label='Cycle 20')
ax3.bar(r4, day_4, yerr = day_4_sd , width=barWidth, edgecolor='white', label='Cycle 30')
ax3.bar(r5, day_5, yerr = day_5_sd , width=barWidth, edgecolor='white', label='Cycle 40')
ax3.bar(r6, day_6, yerr = day_6_sd , width=barWidth, edgecolor='white', label='Cycle 50')

# Add xticks on the middle of the group bars
plt.xlabel('Target Genes')
plt.xticks([r + (2*barWidth) for r in range(len(xlab))], xlab)

# zoom-in / limit the view to different portions of the data
ax1.set_ylim(20, 50)  # outliers only
ax2.set_ylim(4, 10)  # 2nd outliers
ax3.set_ylim(0, 2.5)  # 2nd outliers

# hide the spines between ax1 and ax2 and ax3
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax3.spines['top'].set_visible(False)

ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.get_xaxis().set_visible(False)
ax3.tick_params(labeltop=False)  # don't put tick labels at the top
ax3.xaxis.tick_bottom()

# This looks pretty good, and was fairly painless, but you can get that
# cut-out diagonal lines look with just a bit more work. The important
# thing to know here is that in axes coordinates, which are always
# between 0-1, spine endpoints are at these locations (0,0), (0,1),
# (1,0), and (1,1).  Thus, we just need to put the diagonals in the
# appropriate corners of each of our axes, and so long as we use the
# right transform and disable clipping.

d = .005  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them

kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (-d, +d), **kwargs)  # top-left diagonal
ax2.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax3.transAxes)  # switch to the bottom axes
ax3.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax3.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# Create legend & Show graphic
ax1.legend(loc='upper right', prop={'size':10})
# ax1.set_title('Relative expression of Genes by days')
ax2.set_ylabel('Relative Expression', labelpad = 25)
plt.show()
```

### Function to plot multiple charts 
Since we have multiple datasets to plot, we will put all of the above actions in a function and plot multiple times

```python
def plot_broken_axes(dname, b1, b2, b3, b4, b5, b6):
    
    # get the data 
    set_data = pd.read_csv("data/"+dname+".csv")
    
    # plot parameters
    fig = plt.figure(figsize=(18,8))
    plt.rcParams.update({'font.size': 14})
    fig.subplots_adjust(hspace=0.1)  # adjust space between axes
    # set height ratios for subplots
    gs = gridspec.GridSpec(3, 1, height_ratios=[1, 1, 2]) 
    
    xlab = np.array(set_data['Genes'])
    day_1 = np.array(set_data['1'])
    day_2 = np.array(set_data['2'])
    day_3 = np.array(set_data['3'])
    day_4 = np.array(set_data['4'])
    day_5 = np.array(set_data['5'])
    
    day_1_sd = np.array(set_data['1_sd'])
    day_2_sd = np.array(set_data['2_sd'])
    day_3_sd = np.array(set_data['3_sd'])
    day_4_sd = np.array(set_data['4_sd'])
    day_5_sd = np.array(set_data['5_sd'])

    # set width of bars
    barWidth = 0.1

    # Set position of bar on X axis
    r1 = np.arange(len(xlab))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    r4 = [x + barWidth for x in r3]
    r5 = [x + barWidth for x in r4]


    # Make the plot
    ax1 = plt.subplot(gs[0])
    ax1.bar(r1, day_1, yerr = day_1_sd, width=barWidth, edgecolor='white', label='Day 1')
    ax1.bar(r2, day_2, yerr = day_2_sd,  width=barWidth, edgecolor='white', label='Day 2')
    ax1.bar(r3, day_3, yerr = day_3_sd,  width=barWidth, edgecolor='white', label='Day 3')
    ax1.bar(r4, day_4, yerr = day_4_sd,  width=barWidth, edgecolor='white', label='Day 4')
    ax1.bar(r5, day_5, yerr = day_5_sd,  width=barWidth, edgecolor='white', label='Day 5')

    ax2 = plt.subplot(gs[1], sharex = ax1)
    ax2.bar(r1, day_1, yerr = day_1_sd,  width=barWidth, edgecolor='white', label='Day 1')
    ax2.bar(r2, day_2, yerr = day_2_sd,  width=barWidth, edgecolor='white', label='Day 2')
    ax2.bar(r3, day_3, yerr = day_3_sd,  width=barWidth, edgecolor='white', label='Day 3')
    ax2.bar(r4, day_4, yerr = day_4_sd,  width=barWidth, edgecolor='white', label='Day 4')
    ax2.bar(r5, day_5, yerr = day_5_sd,  width=barWidth, edgecolor='white', label='Day 5')

    ax3 = plt.subplot(gs[2], sharex = ax1)
    ax3.bar(r1, day_1, yerr = day_1_sd,  width=barWidth, edgecolor='white', label='Day 1')
    ax3.bar(r2, day_2, yerr = day_2_sd,  width=barWidth, edgecolor='white', label='Day 2')
    ax3.bar(r3, day_3, yerr = day_3_sd,  width=barWidth, edgecolor='white', label='Day 3')
    ax3.bar(r4, day_4, yerr = day_4_sd,  width=barWidth, edgecolor='white', label='Day 4')
    ax3.bar(r5, day_5, yerr = day_5_sd,  width=barWidth, edgecolor='white', label='Day 5')

    # Add xticks on the middle of the group bars
    plt.xlabel('Target Genes')
    plt.xticks([r + (2*barWidth) for r in range(len(xlab))], xlab)
    
    # zoom-in / limit the view to different portions of the data
    ax1.set_ylim(b1, b2)  # 1st outliers
    ax2.set_ylim(b3, b4)  # 2nd outliers
    ax3.set_ylim(b5, b6)  # main graph
    
    # hide the spines between ax1 and ax2 and ax3
    ax1.spines['bottom'].set_visible(False)
    ax2.spines['top'].set_visible(False)
    ax2.spines['bottom'].set_visible(False)
    ax3.spines['top'].set_visible(False)

    ax1.xaxis.tick_top()
    ax1.tick_params(labeltop=False)  # don't put tick labels at the top
    ax2.get_xaxis().set_visible(False) # delete all the x axes for 2nd chart
    ax3.tick_params(labeltop=False)  # don't put tick labels at the top
    ax3.xaxis.tick_bottom()
    
    d = .005  # how big to make the diagonal lines in axes coordinates
    # arguments to pass to plot, just so we don't keep repeating them

    kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
    ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
    ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

    kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
    ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
    ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

    kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
    ax2.plot((-d, +d), (-d, +d), **kwargs)  # top-left diagonal
    ax2.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

    kwargs.update(transform=ax3.transAxes)  # switch to the bottom axes
    ax3.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
    ax3.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

    # Create legend & Show graphic
    ax1.legend(loc='upper right', prop={'size':11})
#     ax1.set_title('Relative expression of Genes by days')
    ax2.set_ylabel('Relative Expression', labelpad = 25)
    plt.show()
```

## 2. Plotting Heatmaps

### Get the data

```python
# get the data 1
hm_data = pd.read_csv("data/heatmap_data_28.csv")
hm_data = hm_data.set_index('Isoflavonoids')

# get the data 2
hm_data1 = pd.read_csv("data/ic50.csv")
hm_data1 = hm_data1.set_index('Cycles')

# get the data 3
hm_data2 = pd.read_csv("data/inhibition_pct.csv")
hm_data2 = hm_data2.set_index('Cycles')
```
